```python
# 1. 필요한 라이브러리를 불러옵니다.
from datasets import load_dataset
import pandas as pd

# pandas DataFrame의 열 너비를 설정하여 내용이 잘리지 않게 합니다.
pd.set_option('display.max_colwidth', 200)

# 2. 데이터셋을 로드합니다.
# name='all'을 추가하여 'all' 설정을 명시적으로 로드합니다.
print("데이터셋 다운로드를 시작합니다...")
dataset = load_dataset("Hello-SimpleAI/HC3-Chinese", name='all', trust_remote_code=True)
print("다운로드 완료!")

# 3. 로드된 데이터셋의 구조를 확인합니다.
print("\n로드된 데이터셋 구조:")
print(dataset)

# 4. 'train' 데이터를 pandas DataFrame(표)으로 변환합니다.
df = dataset['train'].to_pandas()

# 5. 변환된 DataFrame의 첫 20개 행을 출력합니다.
print("\n데이터셋 상위 20개 샘플 (표 형태):")
df.head(20)
```

```python
# 1. 데이터를 표 형태로 다루기 위해 pandas 라이브러리를 불러옵니다.
import pandas as pd

# 2. GitHub에 있는 원본 파일의 URL을 지정합니다.
file_url = 'https://raw.githubusercontent.com/NLP2CT/NLPCC-2025-Task1/main/data/train.json'

# 3. URL로부터 JSON 파일을 읽어 DataFrame으로 변환합니다.
# lines=True 옵션을 제거하여 표준 JSON 배열로 읽어들입니다.
print(f"{file_url} 에서 데이터를 로드합니다...")
df = pd.read_json(file_url)
print("로드 완료!")

# 4. 데이터의 기본 정보를 확인합니다. (총 행의 수, 열 정보 등)
print("\n데이터프레임 정보:")
df.info()

# 5. 데이터셋의 상위 5개 샘플을 출력하여 모양을 파악합니다.
print("\n데이터셋 상위 5개 샘플:")
df.head()
```
