```python
import gensim.downloader as api
import numpy as np

print("Gensim is ready!")
```

```python
# 간단한 구현 위해 그냥 pre-trained model로
# Load the GloVe model
print("Loading GloVe model...")
glove_model = api.load('glove-wiki-gigaword-100') #100차원, 위키피디아 데이터로 학습한 glove
print("GloVe model loaded successfully!")

# Load the Word2Vec model
print("\nLoading Word2Vec model... (This may take a while)")
word2vec_model = api.load('word2vec-google-news-300') # 300차원, 구글 뉴스데이터로 학습한 word2vec
print("Word2Vec model loaded successfully!")
```

```python
print("## Analogy Test 1: king - man + woman ##\n")

# GloVe
result_glove = glove_model.most_similar(positive=['king', 'woman'], negative=['man'])
print(f"GloVe Result: {result_glove[0]}")

# Word2Vec
result_w2v = word2vec_model.most_similar(positive=['king', 'woman'], negative=['man'])
print(f"Word2Vec Result: {result_w2v[0]}")
```
