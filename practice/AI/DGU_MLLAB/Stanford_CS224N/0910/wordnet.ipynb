```python
# 상위어(Hypernym), 하위어(Hyponym)

dog_synset = wn.synset('dog.n.01')
hypernyms = dog_synset.hypernyms()
hyponyms = dog_synset.hyponyms()

print(f'Hypernyms of "dog.n.01": {hypernyms}')
# canine : 개과, domestic_animal : 가축
print(f'Hyponyms of "dog.n.01": {hyponyms}')
```

```python
# dog.n.01의 root는 뭘까?
root_hypernyms = dog_synset.root_hypernyms()
print(f'Path to the root: {root_hypernyms}') # entity: 독립체, 개체
```

```python
# 의미 유사도 (semantic similarity) : wordnet에선 공통의 hypernym까지 도달하는 경로가 짧을수록 유사함

dog = wn.synset('dog.n.01')
cat = wn.synset('cat.n.01')
car = wn.synset('car.n.01')

# 개와 고양이의 유사도
dog_cat_sim = dog.path_similarity(cat)
print(f'Similarity between "dog" and "cat" : {dog_cat_sim}')

# 개와 자동차의 유사도
dog_car_sim = dog.path_similarity(car)
print(f'Similarity between "dog" and "car" : {dog_car_sim:.4f}')
```
