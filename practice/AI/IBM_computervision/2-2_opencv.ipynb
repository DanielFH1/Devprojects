```python
image = cv2.imread("lenna.png")
print("type: ", type(image))
print("image shape: ", image.shape)

plt.figure(figsize=(5,5))
plt.imshow(image)
plt.show()
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 3. 결과 확인 (비교샷)
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb) # 이제 정상적인 피부색이 나올 거야
plt.title("Correct RGB")

plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap='gray') # 흑백은 cmap='gray' 옵션을 줘야 완벽한 흑백으로 보여
plt.title("Grayscale")
cv2.imwrite("lenna_gray.png",image_gray)

plt.show()
```

```python
blue,green,red = cv2.split(image)
plt.figure(figsize=(10,5))
plt.subplot(1, 3, 1)
plt.imshow(blue, cmap='gray')
plt.title("Blue Channel")

plt.subplot(1, 3, 2)
plt.imshow(green, cmap='gray')
plt.title("Green Channel")

plt.subplot(1, 3, 3)
plt.imshow(red, cmap='gray')
plt.title("Red Channel")

plt.show()
```

```python
image_bgr = cv2.merge([blue,green,red])
image_rgb_merged = cv2.merge([red,green,blue])

plt.subplot(1, 2, 1)
plt.imshow(image_bgr) 
plt.title("Merged (BGR) - Zombie") # matplotlib은 이걸 RGB로 착각하고 그림 -> 파랗게 나옴

plt.subplot(1, 2, 2)
plt.imshow(image_rgb_merged)
plt.title("Merged (RGB) - Human") # 빨강을 빨강 자리에 넣었으니 정상으로 나옴

plt.show()
```

```python
start_row = 150  # 세로 시작 (y1)
end_row   = 400  # 세로 끝   (y2)
start_col = 150  # 가로 시작 (x1)
end_col   = 400  # 가로 끝   (x2)

crop_cv = image[start_row:end_row, start_col:end_col]
crop_cv_rgb = cv2.cvtColor(crop_cv, cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.imshow(crop_cv) 
plt.title("cropped(cv2)") 

plt.subplot(1, 2, 2)
plt.imshow(crop_cv_rgb)
plt.title("cropped(rgb)")

plt.show()
```

```python
flip_horizontal = cv2.flip(image,1)
flip_vertical = cv2.flip(image,0)
flip_both = cv2.flip(image,-1)


plt.figure(figsize=(10,5))
# 원본
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")

# 좌우 반전
plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB))
plt.title("Flip 1 (L-R)")

# 상하 반전
plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(flip_vertical, cv2.COLOR_BGR2RGB))
plt.title("Flip 0 (U-D)")

# 둘 다 반전
plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(flip_both, cv2.COLOR_BGR2RGB))
plt.title("Flip -1 (Both)")

plt.show()
```
