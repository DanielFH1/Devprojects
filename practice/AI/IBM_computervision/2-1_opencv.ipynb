```python
import cv2

image = cv2.imread("lenna.png")
print(type(image))
```

```python
import matplotlib.pyplot as plt

print("Type is: ",image.shape)

plt.figure(figsize=(10,10))
plt.imshow(image)
plt.show()
```

```python
# As you can see above it looks bizarre compare to PIL
# This is because opencv consider color class as B-G-R order, and matplotlib consider color class as R-G-B
# So in opencv, first we have to convert the order

new_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(new_image)
plt.show()
```

```python
# Make Grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print("Gray scale image shape: ", image_gray.shape) # (512,512) 

plt.figure(figsize=(10,10))
plt.imshow(image_gray, cmap='gray')
plt.show()
```

```python
im_gray = cv2.imread('barbara.png',cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10,10))
plt.imshow(im_gray, cmap='gray') 
# If there is no cmap='gray', since the shape is 2D(512,512) matplotlib just fill the color randomly
plt.show()
```

```python
# Slicing Channels : Remember opencv has the order of B G R

baboon = cv2.imread('baboon.png')

blue = baboon[:,:,0]
green = baboon[:,:,1]
red = baboon[:,:,2]

fig,axes = plt.subplots(1,3,figsize=(15,5))

axes[0].imshow(blue, cmap='gray')
axes[0].set_title("Blue channel")

axes[1].imshow(green, cmap='gray')
axes[1].set_title("Green channel")

axes[2].imshow(red, cmap='gray')
axes[2].set_title("Red channel")
```

```python
baboon_red = baboon.copy()

baboon_red[:,:,0] = 0
baboon_red[:,:,1] = 0

plt.figure(figsize=(10,10))

plt.imshow(cv2.cvtColor(baboon_red, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
baboon = cv2.imread('baboon.png')
baboon_rgb = cv2.cvtColor(baboon, cv2.COLOR_BGR2RGB)

baboon_blue = baboon_rgb.copy()

baboon_blue[:,:,0]=0
baboon_blue[:,:,1]=0

plt.figure(figsize=(10,10))
plt.imshow(baboon_blue)
plt.show()
```
