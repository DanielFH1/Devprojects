```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
```

```python
print(x_train.shape)
x_train.shape[0]
```

```python
plt.scatter(x_train,y_train,marker='o',c='b')
plt.title("Housing prices")
plt.ylabel("price")
plt.xlabel("Size of house")
plt.show()
```

```python
def compute_model_output(x,w,b):
    """
    단일 변수 선형 회귀 모델의 예측값을 계산
    x : input(scalar or matrix)
    w : weight
    b : bias
    """
    m=x.shape[0] # 입력 샘플 개수
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
    return f_wb
```

```python
w = 200
b = 100
tmp_f_wb = compute_model_output(x_train, w,b)
print(tmp_f_wb)
```

```python
#산점도로 실제 데이터 표시
plt.scatter(x_train,y_train,marker="o",c="b",label="Actual Values")

#모델 예측값
plt.plot(x_train,tmp_f_wb,c="r",label="our prediction")

plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()  # 범례 표시
plt.show()
```

```python
#여러모델을 한번에 시각화

w_array = [200,150,120]
b_array = [1000,50,250]

# 실제 데이터 산점도
plt.scatter(x_train,y_train, marker="o", c= "b", label = 'actual values')

#여러 (w,b)조합마다의 예측선
for w,b in zip(w_array,b_array):
    f_wb = compute_model_output(x_train,w,b)
    plt.plot(x_train,f_wb,label=f"w={w},b={b}")

plt.title("Housing Prices: Various data sets")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
